```python
def run_scored(batsman_name):
    return delivery[delivery['batsman']==batsman_name].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
run_scored('S Dhawan')
```

```python
def run_scored_against(batsman_name):
    return delivery[delivery['batsman']==batsman_name].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

# Overall strike rate

```python
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
df = pd.read_csv('test.csv')
```

```python
df
```

```python
df.drop(['PassengerId','Ticket','Name'],axis=1,inplace=True)
df.head()
```

A single index

Using the ‘Sex’ column as the index for now:

```python
table1 = pd.pivot_table(data=df,index=['Sex'])
table1
```

```python
table1.plot(kind='bar')
```

Multiple index index


You can even use more than one feature as an index to group your data. This increases the level of granularity in the resultant table and you can get more specific with your findings:

```python
# multiple indexes

table2 = df.pivot_table(index=['Sex','Pclass'])
table2
```

```python
table2.plot(kind='bar')
```

2.2 Different aggregation function for different features


Using multiple indexes on the dataset enables us to concur that the disparity in ticket fare for female and male passengers was valid across every Pclass on Titanic.

Using np.mean() for the ‘Age’ feature and np.sum() for the ‘Survived’ feature;

```python
# different aggregate functions

table3 = df.pivot_table(index=['Sex','Pclass'],
                        aggfunc={'Age':np.mean})
table3
```

```python
table3.plot(kind='bar')
```

```python
# display null values

table6 = df.pivot_table(index=['Sex','Pclass'],
                        columns=['Embarked'],
                        values=['Age'],
                        aggfunc=np.mean)
table6
```

```python
table6.plot(kind='bar')
```

```python
table7 = pd.pivot_table(df,index=['Sex','Pclass'],
                        columns=['Embarked'],
                        values=['Age'],
                        aggfunc=np.mean,
                        fill_value=np.mean(df['Age']))
table7
```

```python
table7.plot(kind='bar')
```
